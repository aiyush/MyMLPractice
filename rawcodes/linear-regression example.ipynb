```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
df = pd.read_csv("data/train.csv")
print(df)
x,y = df.loc[:400,'x'],df.loc[:400,'y']
#x,y = data[['x']],data[['y']]
print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#Learning Rate
alp = 0.00001

#Stopping criteria
ep = 1e-06

#Choose an error value
err = 1

#Total number of data
m = 400

#Choosing w0 and w1 randomly 
w0,w1 = random.rand(),random.rand()
#print(f"w0 = {w0},w1 = {w1}")

yh = (w1*x)-w0
S = y-yh
#print(f"yh = {yh}, S = {S}")

#Initial cost function
J = (0.5/m)*(np.sum(S)**2)
#print(f"Sum of J is: {J}")
#print(J)

Jloop = 0

while(err>ep):
    yhn = w1*x + w0
    yhnsum = np.sum(yhn)
    
    DJ0 = yhn-y
    dw0 = -alp*np.sum(DJ0)/m
    
    DJ1 = (yhn-y)*x
    dw1 = -alp*np.sum(DJ1)/m
    
    w0 += dw0
    w1 += dw1
    
    yhupdated = w1*x - w0
    sumupdated = np.sum(y-yhupdated)
    
    Jnew = 0.5/m*(sumupdated**2)
    
    err = abs(Jnew - Jloop)
    
    Jloop = Jnew
    
    print(err)
    
    plt.subplot(2,1,1)
    plt.plot(x,y,'o',x,yhupdated,'r');
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()  
print(f"The Value of w0 is: {w0}")
print(f"The value of w1 is: {w1}")

```
